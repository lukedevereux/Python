# **Scatter Plots**

Let's try a different form of plot. A scatter plot. Let's also create some new data. We will create some imaginary data for budget spent and for impressions (these are made up for illustrative purposes)

```python
Budget = [50, 10, 100, 23, 150, 65, 200, 76, 32, 130, 67, 80, 180, 190, 50,10, 99,195,123,140] #creates an imaginary budget variable
```

```python
Impressions = [100,20,200,40,300, 120, 400, 140, 60, 270, 135, 160, 390, 420, 120,15,170,410,240,286] #this creates an impressions variable
```

Now that we have our fake data. Let's plot this on a graph.

```python
fig, Secgraph = plt.subplots() #fig helps us create background. We've named Sec graph (basically secondgraph)
Secgraph.scatter(Budget, Impressions) # we use scatter here and specify what we want it to plot.
```

Now let's add some bits to this. Titles and changing color of graphs slightly like how we did last time.

```python
fig, Secgraph = plt.subplots() #fig helps us create background. We've named Sec graph (basically secondgraph)
Secgraph.scatter(Budget, Impressions, s=50, facecolor='red', edgecolor='black') #changed the size and color of parts
Secgraph.set_xlabel('Budget', fontsize=15) #added in the Budget title and fot zsize
Secgraph.set_ylabel('Impressions', fontsize=15) #added in Impressions title and font size
Secgraph.set_title('Relationship between budget and impressions', fontsize=17) #added in the main title and font size
plt.show() #shows the graph
```

Let's say we had a third piece of data to map on here and we wanted to represent this by color. For example, maybe the number of clicks. Lets first create this data.

```python
Clicks = [300,8,160,33,260, 100, 380, 111, 41, 215, 15, 124, 363, 398, 100,4,148,390,217,259]
```

Now lets put this into the graph using colour to denote this. We will also just change the size of the points to help make more visible.

```python
fig, Secgraph = plt.subplots(figsize=(10, 7)) #fig helps us create background. We've named Sec graph (basically secondgraph)
pc = Secgraph.scatter(Budget, Impressions, c=Clicks, cmap='RdBu_r', s=90) #changed the size and color. we've named this something 'pc' for later
Secgraph.set_xlabel('Budget', fontsize=15) #added in the Budget title and font zsize
Secgraph.set_ylabel('Impressions', fontsize=15) #added in Impressions title and font size
Secgraph.set_title('Relationship between budget and impressions', fontsize=17) #added in the main title and font size
fig.colorbar(pc, ax=Secgraph, extend='both') #adds the color bar pc is the name we gave above to graph

plt.show() #shows the graph
```

# Categorical Variables and Graphs

We've looked at numerical data so far.Let's now look at some categorical data plotting. Let's imagine we have some different data about different platforms e.g instagram,fb, Tiktok etc.

Lets create some fake data. We will do this through a slightly different method using a 'dictionary'.

```python
data = {'instagram': 39, 'facebook': 25, 'Tiktok': 45, 'Twitter': 25} #create our data we've just added random ones
names = list(data.keys()) # creates a variable called names and makes this a list involving data variable and its keys
values = list(data.values()) #creates a variable called values and makes this a list involving data variable and its values

fig, axs = plt.subplots(1, 3, figsize=(15, 7), sharey=True) #create out figure and set size
axs[0].bar(names, values) #creates a bar graph
axs[1].scatter(names, values) #creates a scatter
axs[2].plot(names, values) #creates another line plot
fig.suptitle('Categorical Plotting in different ways')
```

Lets just make an example of one of these. We will use the bar chart example. We will first create a 'dictionary' that links a platform name with some data about it. So a key and a value. e.g. 'Instagram: 39' tells us the platform is Instagram and the value is 39.

```python
#Let's create the data first
PlatformData = {'instagram': 39, 'facebook': 25, 'Tiktok': 45, 'Twitter': 25} #create our data using a dictionary we've just added random ones and name it PlatformData
Platforms = list(PlatformData.keys()) # this creates new variable 'names' and links it to the keys/categories above
values = list(PlatformData.values()) #this does the same as above but creates something called values, and makes it the values
```

Now lets add some labels to this. We do this using Thirdgraph.set_xlabel() as well as .set_ylabel() and .set_title()





```python
fig, Thirdgraph = plt.subplots(figsize=(12, 5)) #creates our figure to play with
Thirdgraph.bar(Platforms, values, color ='green') #creates the graph and we have used color to add in the color we want.
Thirdgraph.set_xlabel('Platform', fontsize=15) #added in the Budget title and fot zsize
Thirdgraph.set_ylabel('Clicks', fontsize=15) #added in Impressions title and font size
Thirdgraph.set_title('No. of clicks per platform', fontsize=17) #added in the main title and font size
plt.show()#show our graph
```

Okay lets see if we can make this include two types of data, so lets imagine we were comparing Campaign 1 and Campaign 2 again.

```python
#Lets create the data we need for this and store it in something called Channel
Channel = ['instagram', 'facebook', 'Tiktok', 'Twitter'] #this creates our labels
#now lets create our data
Cam1 = [43,16,87,33] #fictional number of clicks for campaign 1
Cam2 = [12,6,55,25] #fictional number of clicks for campaign 2
```

Now lets make our graph

```python
x = np.arange(len(Channel))  # we use numpy here to help create even arrange spacing based on length of channels
width = 0.35  # the width of the bars

fig, Camgraph = plt.subplots(figsize=(12, 5)) #creates our figure like before
rects1 = Camgraph.bar(x - width/2, Cam1, width, label='Campaign 1', color = 'green') #creates bar for cam 1 stores it as rects1
rects2 = Camgraph.bar(x + width/2, Cam2, width, label='Campaign 2', color = 'orange') #creates bar for cam2 and stores it as rects2

# Add some text for labels, title and custom x-axis tick labels, etc.
Camgraph.set_ylabel('Clicks')
Camgraph.set_xlabel('Platform')
Camgraph.set_title('Scores by Campaign')
Camgraph.set_xticks(x, Channel)
Camgraph.legend()

Camgraph.bar_label(rects1, padding=3) #adds the platform labels
Camgraph.bar_label(rects2, padding=3) #adds platform labels
```

```python


labels = ['G1', 'G2', 'G3', 'G4', 'G5']
men_means = [20, 34, 30, 35, 27]
women_means = [25, 32, 34, 20, 25]

x = np.arange(len(labels))  # the label locations
width = 0.35  # the width of the bars

fig, ax = plt.subplots()
rects1 = ax.bar(x - width/2, men_means, width, label='Men')
rects2 = ax.bar(x + width/2, women_means, width, label='Women')

# Add some text for labels, title and custom x-axis tick labels, etc.
ax.set_ylabel('Scores')
ax.set_title('Scores by group and gender')
ax.set_xticks(x, labels)
ax.legend()

ax.bar_label(rects1, padding=3)
ax.bar_label(rects2, padding=3)

fig.tight_layout()

plt.show()
```
